```python
 f2 = pd.read_csv('C:/Users/197212/Documents/Learnings/SuperLeague/TimeSeries/Test-Data.csv', parse_dates=["date"], index_col="date")
 f2.head()
```

```python
#Run the below cell to filter the closing price related to stock 'AA'

closeTS = f2[(f2.stock == 'AA')].close.str.replace('$',' ').astype(float)
```

```python
#How many observations have you seen from Jan 1, 2011 to March 31, 2011?
closeTS
```

```python
#Perform a monthly resample/downsample of the time series. What is the maximum value for February?
closeTS.resample('1m').max().head()
```

```python
#Perform a daily resample/upsample of the data. Do interpolation to fill the data. 
#What is the value for Jan 12, 2011?
closeTS.resample('1d').interpolate().head(10)
```

```python
#Perform a daily resample/upsample of the data. Do a forward filling of the missing values with limit of 2. 
#What is the value for Jan 16,2011?
closeTS.resample('1d').ffill(limit=2).head(10)
```

```python
#Perform a monthly resample/downsample of the time series. What is the minimum value for May?
closeTS.resample('1m').min().head()
```
