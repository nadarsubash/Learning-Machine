## HANDSON - Sine Function

```python
fig = plt.figure(figsize=(12,3))
ax = fig.add_subplot(111)
t = np.linspace(0.0,2.0,num=200)
v = np.sin(2.5*np.pi*t)
plt.plot(t,v,color='red',label='sin(t)')
plt.xlabel('Time (seconds)')
plt.ylabel('Voltage (mV)')
plt.title('Sine Wave')
plt.xlim(0,2)
plt.ylim(-1,1)
plt.xticks(np.arange(0, 2.1, 0.2)) 
plt.yticks(np.arange(-1, 2, 1))
plt.legend()
plt.grid(linestyle='--')

plt.show()
```

## HANDSON - Polynomial

```python
fig = plt.figure(figsize=(12,3))
ax = fig.add_subplot(111)

x = np.linspace(0.0,5.0,num=20)
y1 = x
y2 = x**2
y3 = x**3

plt.plot(x,y1,color='red',label='y = x',marker='o')
plt.plot(x,y2,color='green',label='y = x**2',marker='s')
plt.plot(x,y3,color='blue',label='y = x**3',marker='^')
plt.plot()

plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Linear, Quadratic, & Cubic Equations')
plt.legend()

plt.show()
```

## BAR PLOT
[Tutorial](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)

```python
from matplotlib import pyplot
fig = pyplot.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales',
      xlabel='Quarter', ylabel='Sales (in millions)')
quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]
pyplot.bar(quarters, sales_2017, color='red', width=0.6, edgecolor='black')
#ax.set_xticks(quarters)
#ax.set_xticklabels(['Q1-2017', 'Q2-2017', 'Q3-2017'])
pyplot.xticks(quarters,('Q1-2017', 'Q2-2017', 'Q3-2017'))
pyplot.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales',
      xlabel='Quarter', ylabel='Sales (in millions)')
quarters = [1, 2, 3]
x1_index = [0.8, 1.8, 2.8]; x2_index = [1.2, 2.2, 3.2]
sales_2016 = [28831, 30762, 32178]; sales_2017 = [25782, 35783, 36133]
ax.bar(x1_index, sales_2016, color='yellow', width=0.4, edgecolor='black', label='2016')
ax.bar(x2_index, sales_2017, color='red', width=0.4, edgecolor='black', label='2017')
ax.set_xticks(quarters)
ax.set_xticklabels(['Q1', 'Q2', 'Q3'])
ax.legend()
plt.show()
```

## SUBPLOT
[Tutorial](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplot.html)

subplot(nrows, ncols, index)<br>
'index' is the position in a virtual grid with 'nrows' and 'ncols'<br>
'index' number varies from 1 to `nrows*ncols`.

##### Creating a simple 2 x 2 plots in a single figure

```python
fig = plt.figure(figsize=(10,8))
axes1 = plt.subplot(2, 2, 1, title='Plot1')
axes2 = plt.subplot(2, 2, 2, title='Plot2')
axes3 = plt.subplot(2, 2, 3, title='Plot3')
axes4 = plt.subplot(2, 2, 4, title='Plot4')
plt.show()
```

##### Manipulating the plot matrix
Now let's try to create a figure with three subplots, where the first subplot spans all columns of first row.

```python
fig = plt.figure(figsize=(10,8))
axes1 = plt.subplot(2, 2, (1,2), title='Plot1')
axes2 = plt.subplot(2, 2, 3, title='Plot2')
axes3 = plt.subplot(2, 2, 4, title='Plot3')
plt.show()
```

Now let's try to remove the ticks in the plots

```python
fig = plt.figure(figsize=(10,8))
axes1 = plt.subplot(2, 2, (1,2), title='Plot1')
axes1.set_xticks([]); axes1.set_yticks([])
axes2 = plt.subplot(2, 2, 3, title='Plot2')
axes2.set_xticks([]); axes2.set_yticks([])
axes3 = plt.subplot(2, 2, 4, title='Plot3')
axes3.set_xticks([]); axes3.set_yticks([])
plt.show()
```

## Subplot using GRIDSPEC
Specifies the geometry of the grid that a subplot can be placed in. <br>
The location of grid is determined by similar way as the SubplotParams.<br>
The number of rows and number of columns of the grid need to be set. Optionally, the subplot layout parameters (e.g., left, right, etc.) can be tuned.

[Tutorial](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.gridspec.GridSpec.html)

##### sample gridspec
recreating above plot using gridspec

```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,8))
gd = gridspec.GridSpec(2,2)
axes1 = plt.subplot(gd[0,:],title='Plot1')
axes1.set_xticks([]); axes1.set_yticks([])
axes2 = plt.subplot(gd[1,0],title='Plot2')
axes2.set_xticks([]); axes2.set_yticks([])
axes3 = plt.subplot(gd[1,-1],title='Plot3')
axes3.set_xticks([]); axes3.set_yticks([])
plt.show()
```
